### Task2
<a id="task2"></a> 

```python
#read csv file for task2
task2 = pd.read_csv('csv_tasks/task2.csv',sep="^",skiprows = 4)
task2
```

```python
# As we can see from data in task2, country is a nominal attribute. 
# Series is a Ratio-scaled attribute, because of zero-point, values have order and measured on a scale of equal-sized units.
```

```python
# create a barplot with counts of series per country
sns.set_style("whitegrid")
plt.figure(figsize=(20,8))
plt.title('Barplot with counts of series per country')
sns.barplot(x='country', y='series', data=task2, palette='tab10')
plt.xticks(rotation=75);
```

United States have the most series in number than other countries

```python
# Create a pieplot
plt.figure(figsize=(14,4))
plt.pie(task2.series, labels=task2.country, radius = 3, autopct='%1.2f%%');
```

We can see, that United States have more than 1/3 of series. Then, we can see United Kingdom with 11.35%.

```python
# Boxplot with series values
plt.figure(figsize=(3,6))
plt.title('Boxplot with the series values')
sns.boxplot(y='series',data=task2);
```

Most countries have series lower than 50, and over 50 there are 7 countries

### Task3
<a id="task3"></a> 

```python
#read csv file for task3
task3 = pd.read_csv('csv_tasks/task3.csv',sep="^",skiprows = 4)
task3
```

```python
# As we can see from data in task3, genre is a nominal attribute. 
# Productions is a Ratio-scaled attribute, because of zero-point, values have order and measured on a scale of equal-sized units.
```

```python
# create a barplot with counts of productions per genre
sns.set_style("whitegrid")
plt.figure(figsize=(14,4))
plt.title('Barplot with counts of productions per genre')
sns.barplot(x='genre', y='productions', data=task3, palette='tab10')
plt.xticks(rotation=75);
```

International Movies have the most productions and then we can see dramas, comedies and international TV shows.

```python
# Create a donut chart
plt.figure(figsize=(14,4))
plt.pie(task3.productions, labels=task3.genre, radius = 3, autopct='%1.2f%%')
# add a circle at the center to transform it in a donut chart
circle=plt.Circle( (0,0), 1, color='white')
plt.gcf().gca().add_artist(circle);
```

International Movies have the most productions with 14.1% and then we can see dramas with 11.87%. All the other values are lower than 10%.

```python
# create a boxplot with productions
plt.figure(figsize=(3,6))
plt.title('Boxplot with productions of genre')
sns.boxplot(y='productions',data=task3);
```

Most of the values are lower than 750, and we can see 4 values over them. Maybe these values are international movies, dramas, comedies and international TV shows.

### Task4
<a id="task4"></a> 

```python
#read csv file for task4
task4 = pd.read_csv('csv_tasks/task4.csv',sep="^",skiprows = 4)
task4
```

```python
# As we can see from data in task4, actor is a nominal attribute. 
# Productions is a Ratio-scaled attribute, because of zero-point, values have order and measured on a scale of equal-sized units.
```

```python
# create a barplot with counts of productions per actor
sns.set_style("whitegrid")
plt.figure(figsize=(10,3))
plt.title('Barplot with counts of productions per actor')
sns.barplot(x='actor', y='productions', data=task4, palette='tab10')
plt.xticks(rotation=60);
```

```python
# Create a donut chart
plt.figure(figsize=(14,4))
plt.pie(task4.productions, labels=task4.actor, radius = 2.5, autopct='%1.2f%%')
# add a circle at the center to transform it in a donut chart
circle=plt.Circle( (0,0), 1, color='white')
plt.gcf().gca().add_artist(circle);
```

From the donut chart, we see that all the actors are very close, with the first one with 7.19% and the last one with 3.92%, which difference is near to 3%, which is not too much.

```python
# create a boxplot with productions
plt.figure(figsize=(3,6))
plt.title('Boxplot with productions of actors')
sns.boxplot(y='productions',data=task4);
```

We can see that the median is near to 22, and the values are from 18 to 26 from Q1 to Q3, and 33 is the upper extreme.

### Task5
<a id="task5"></a> 

```python
#read csv file for task5
task5 = pd.read_csv('csv_tasks/task5.csv',sep="^",skiprows = 4)
task5
```

```python
# As we can see from data in task5, actor and genre are nominal attributes. 
# Works is a Ratio-scaled attribute, because of zero-point, values have order and measured on a scale of equal-sized units.
```

```python
# create a countplot with the genre attribute
sns.set_style("whitegrid")
plt.figure(figsize=(10,7))
plt.title('Countplot with the genre attribute')
sns.countplot(y='genre',data=task5);
```

In this plot we can see that, Dramas are the most common genre for more than 5000 actors. Then, we have comedies with 4000 and around to 3000 are international tv shows and action & adventures. All the other genres are lower than 2000.

```python
# subplot with boxplot with works of actor in the most common genre of them
plt.figure(figsize=(25,7))
plt.title('Boxplots with works of actor in the most common genre')
sns.boxplot(x='genre', y='works',data=task5)
plt.xticks(rotation=70);
```

We can see that international movies values are very sparse, and 15 values are over Q3 percentile, which is equal to 8 and extreme upper is equal to 30. Most of the genres have extreme upper values lower than 10 works per actor. Of course, the median in most of the genres are around to 1.

```python
# stripplot with works of actor in the most common genre
plt.figure(figsize=(20,7))
plt.title('Stripplot with works of actor in the most common genre')
sns.stripplot(x='genre', y='works',data=task5)
plt.xticks(rotation=70);
```

We can see that most values of each genre are lower than 5 and 6 genres have lots of values more than 5, 
which are international movies, action & adventure, dramas, anime series, children & family movies and comedies.
